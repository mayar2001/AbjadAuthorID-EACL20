```python
import transformers
from transformers import Trainer, TrainingArguments
from datasets import Dataset
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from arabert.preprocess import ArabertPreprocessor
```

```python
val_df = pd.read_excel('/content/gdrive/MyDrive/Colab Notebooks/AbjadAuthIden/AuthorshipClassficiationVal.xlsx')
```

```python
df=pd.read_excel('/content/gdrive/MyDrive/Colab Notebooks/AbjadAuthIden/AuthorshipClassficiationTrain.xlsx')
```

```python
d = dict(zip(df.author.unique(), range(0,len(df.author.unique()))))
val_df['author']=val_df['author'].map(d)
df['author']=df['author'].map(d)
```

```python
d
```

```python
model_name = "aubmindlab/bert-base-arabertv2"
arabert_prep=ArabertPreprocessor(
    model_name=model_name,
    strip_tashkeel=True,
    strip_tatweel=True,
)
```

```python
val_df["text_in_author_style"] = val_df["text_in_author_style"].apply(arabert_prep.preprocess)
```

```python
df["text_in_author_style"] = df["text_in_author_style"].apply(arabert_prep.preprocess)
print(df)
```

```python
train_dataset = Dataset.from_pandas(df)
val_dataset = Dataset.from_pandas(val_df)
```

```python
train_dataset = train_dataset.rename_column("author", "labels")
val_dataset = val_dataset.rename_column("author", "labels")
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=21)
```

```python
def tokenize_fn(batch):
    return tokenizer(
    batch["text_in_author_style"],
    max_length=512,
    truncation=True,
    stride=128,
)
```

```python
train_dataset = train_dataset.map(tokenize_fn,batched=True)
val_dataset = val_dataset.map(tokenize_fn,batched=True)
```

```python
val_dataset
```

```python
train_dataset = train_dataset.remove_columns(["text_in_author_style","id"])
val_dataset = val_dataset.remove_columns(["text_in_author_style","id"])
```

```python
# Rename the 'author' column to 'labels' for the Trainer to correctly identify them
train_dataset.set_format(type="torch", columns=['labels', 'input_ids', 'token_type_ids', 'attention_mask'])
val_dataset.set_format(type="torch", columns=['labels', 'input_ids', 'token_type_ids', 'attention_mask'])
```

```python
from transformers import TrainingArguments
import torch
import accelerate

training_args = TrainingArguments(
    output_dir="./AbjadAuthIden1",
    per_device_train_batch_size=8,
    num_train_epochs= 3,
    learning_rate=2e-5,
    #learning_rate=1e-5,

    do_eval=True,
    logging_steps=100,

    save_steps=5000,
    save_total_limit=2,
    eval_strategy="steps",
    eval_steps=500,
    optim="adamw_torch",
    # keep only last 3 checkpoints
    load_best_model_at_end=True,   # restore best checkpoint
    metric_for_best_model="f1",
    greater_is_better=True,
    report_to=[]
)
```

```python
import numpy as np
import sklearn
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)

    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds, average="weighted")
    }

```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics

)
```

```python
trainer.train()
```

```python
test=pd.read_excel('/content/gdrive/MyDrive/Colab Notebooks/AbjadAuthIden/PublicDataFinalPhaseAbjadAuthID.xlsx')
```

```python
test["text_in_author_style"] = test["text_in_author_style"].apply(arabert_prep.preprocess)
test = Dataset.from_pandas(test)
```

```python
def tokenize_fn_test(batch):
    tokenized = tokenizer(
        batch["text_in_author_style"],
        max_length=512,
        truncation=True,
    )
    return tokenized
```

```python
test = test.map(
    tokenize_fn_test,
    batched=True,
)
```

```python
import numpy as np
from scipy.special import softmax

# preds: raw logits from trainer.predict
# shape = (num_samples, 21)
predictions_test = trainer.predict(test)
logits = predictions_test.predictions
probs = softmax(logits, axis=1)  # convert logits -> probabilities

num_classes = probs.shape[1]
assert num_classes == 21  # sanity check

# Default: threshold = 0.5 for all classes (you can tune later)
thresholds = [0.5] * num_classes

pred_labels = []
for p in probs:
    # find classes where probability > threshold
    above = [i for i in range(num_classes) if p[i] > thresholds[i]]
    if len(above) == 0:
        # fallback to highest probability if no threshold is passed
        pred_labels.append(np.argmax(p))
    else:
        # pick first class that passed threshold
        pred_labels.append(above[0])
```
